# Introducción a Matplotlib gráfico de líneas y dispersión
En este caso veremos la sintaxis a seguir para crear dos gráficos básicos
- Lineal - De linea - plot, gráfico continuo.
- Scatter - Gráfico de dispersión, sirve para ver si hay o no relación entre variables, existe una correlación o no.

En este caso usaremos un MÓDULO de matplotlib "pyplot", en uso tambien de arrays de Numpy 

# Estructura básica para crear un gráfico.
- Define tus dos variables, cuida que seán del mismo tamaño. En este caso las crearemos con Numpy. Hicimos un poco dinámico la variable Y solo para recordar conceptos de Numpy, en este caso el uso de su módulo Random
- Configura el tamaño del gráfico (estandar) con figure(figsize=(,))
- Crea el TIPO de gráfico - type(2 variables PRIMERO VA X, LUEGO Y, parámetros propios de marker (forma en la que se van a marcar cada uno de los datos cuando este en x y en y), color, etc.)
- Añade las etiquetas que ayudan a su visualización - title(''), xlabel('') y ylabel('')
- Muestralo - show.()

```python
import matplotlib.pyplot as plt
import numpy as np

month = np.array(["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"])
sales = np.random.randint(50, 200, 6) #Usamos el MÓDULO DE np llamado Random, en especifico usamos la función de valores aleatorios enteros y como PARÁMETROS, los primeros dos son el rango de los valores que me va a devolver y el tercero es el tamaño, cuantos valores me dará.

#Configurar el tamaño del gráfico
plt.figure(figsize=(8,6))

#Creamos el gráfico
plt.plot(month, sales, marker='o', color='blue')

plt.title('Ventas mensuales de un producto')
plt.xlabel('Meses')
plt.ylabel('Venta')
```

```python
# Configuramos el tamaño del gráfico
plt.figure(figsize=(8,6))

#Defino mis variables para ver si existe correlación entre ellas, en este caso será el puntaje obtenido de un solo estudiante y sus horas estudiadas.
hours = [2,3,5,6,7,10]
exam_score = np.random.randint(2,6,6)
#TIENE QUE TENER LA MISMA DIMENSIÓN, NÚMERO DE ENTRADAS.

#Defino el tipo de gráfico que quiero
plt.scatter(hours, exam_score, color='green')

plt.title('Relación entre horas estuduadas y el puntaje obtenido')
plt.xlabel('Horas')
plt.ylabel('Puntaje')
```

# Parámetros de nuestro gráfico seleccionado
Tanto para scattter y plot (gráfico de líneas) podemos utilizar:
- marker = '' - Parámetro que define la forma en la que se van a marcar cada uno de los datos cuando este en x y en y
- label = 'Título al grafico' - Leyenda, se DEBE DE COMPLETAR, plt.legend()

Parámetros extra para el gráfico de líneas:
- linewidth = 2
- linestyle = '--' - Queremos que lo separe por líneas o bien podría ser '-' para una linea continua

Si se añaden al de scatter simplemente los omite, no los utiliza.

```python
plt.figure(figsize=(8,6))

hours = [2,3,5,6,7,10]

#Estudiante 1
exam_score_1 = [55,34,70,67,90,85]

#Estudiante 2
exam_score_2 = [60,65,70,40,75,90]


# Gráficos de dispersión de dos estudiantes.
plt.scatter(hours, exam_score_1, marker='o', color='green', linestyle = '-', label = 'Estudiante 1')
plt.scatter(hours, exam_score_2, marker='s', color='red', linestyle = '--', label = 'Estudiante 2' )

plt.title('Relación entre horas estuduadas y el puntaje obtenido de dos estudiantes')
plt.xlabel('Horas')
plt.ylabel('Puntaje')
```

# Gráfico de Barras y Diagrama de Pastel
Sirven para representar datos categóricos de manera  visual y comprenisible 
- Los de barra son ideales para comparar cantidades entre diferentes categorias
- Los de pastel para mostrar la proporción de cada categoría pero en un todo.

```python
print('Gráfico de barras...\n')
plt.figure(figsize=(8,6))

categories = ['Product A', 'Product B', 'Product C']
sales = [120,120,90]

plt.bar(categories, sales, color='skyblue', label='Ventas mensuales')

plt.title('Ventas de productos en un mes')
plt.xlabel('Productos')
plt.ylabel('Ventas')

plt.legend()
plt.show()

```

Veamos que además que podemos darle un sentido horizontal y tomar como referencia algún valor, por ejemplo para este caso podemos tomar la barra más grande.

Lo siguiente por añadir tambien lo podemos implementar en los otros gráficos vistos...

Primero agamos referencia a un valor y personalizarlo ya sea con texto o señalandolo.

plt.annotate('') que recibe como parámetros
- Título o nombre.
- xy que corresponde a dar la coordenada de los valores que cumplen la característica que queremos señalar.
- xytext - Podemos añadir además  una flecha para este caso que apunte a XY dandole el inicio de la flecha... 
=('Product C', 160), le paso las coordenadas de donde inicia la flecha, apuntando hacia el parámetro XY
- arrowprops se le asigna un DICCIONARIO con dos parámetros, estos deben de estar dentro de una tupla.
=dict(facecolor='Black', shrink=0.05)

```python
plt.figure(figsize=(8,6))

categories = ['Product A', 'Product B', 'Product C']
sales = [120,150,90]

plt.bar(categories, sales, color='skyblue', label='Ventas mensuales')

plt.annotate('Valor Máximo', xy=('Product B', 150), xytext=('Product C', 160), arrowprops=dict(facecolor='Red', shrink=0.05))

plt.title('Ventas de productos en un mes')
plt.xlabel('Productos')
plt.ylabel('Ventas')

plt.legend()
plt.show()
```

Para el sentido horizontal solo hay que indicar que queremos ese tipo de gráfica, es uno diferente al de bar, en este caso es barh

Además de que los ejes se invierte, primero se asigna el eje Y y luego el X, no se cambian los nombres de los ejes, si no que la posición en la que son escritos.

```python
plt.figure(figsize=(8,6))

categories = ['Product A', 'Product B', 'Product C']
sales = [120,150,90]

plt.barh(categories, sales, color='skyblue', label='Ventas mensuales')

plt.title('Ventas de productos en un mes')
plt.ylabel('Productos')
plt.xlabel('Ventas')

plt.legend()
plt.show()
```

# Diagramas de pastel
Su tamaño es proporcional a su valor de cada categoria, es recomendable su uso cuando se tiene pocas categorias con diferencias claras en las porporciones.

Es complejo personalizarlo pero esto es lo que debe de recibir...

Al definir el tipo de gráfico que queremos, que en este caso será pie, recibe como parámetros:
- PRIMERO van los VALORES
- Luego con labels que van a parecer en cada división del pastel, en este caso los productos
- autopct - Va a dar proporción a cada una de las partes en el diagrama. '1.1f%%'. ESTO HACE REFERENCIA A LOS VALORES QUE SE MUESTRAN DENTRO DEL GRÁFICO EN FORMA DE PORCENTAJE.
- startangle - En que ángulo debe de empezar a hacer las divisiones = 140
- colors - En este caso se pasarán 3 EN FORMA DE LISTA, se deben de pasar la cantidad de colores con respecto a la cantidad de valores que tengas en tu variable

Luego debemos de asegurar que se nos este dando un ciruclo y no alguna otra figura como un elipse, esto ya no es un parámetro de la figura.
plt.axis('Equal')


```python
plt.figure(figsize=(8,6))

categories = ['Product A', 'Product B', 'Product C']
sales = [120,150,90]

plt.pie(
    sales, 
    labels=categories, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=['gold', 'lightcoral', 'orange'])

plt.axis('equal')

plt.title('Ventas de productos en un mes')

plt.legend()
plt.show()
```

# Boxplots 
Diagramas de caja son gráficos que resumen la distribución de datos a través de sus quartiles
Permiten identificar la mediana, la dispersión y los valores atípicos en un conjunto de datos.

Breve explicación en la descripción de la clase.

En este caso usaremos una “función” del módulo random de numpy llamada seed()…
Establecemos una semilla, cada que la que ejecutemos iniciará desde el mismo lugar, entonces cuando ejecutemos la parte aleatoria vamos a mantener la misma informacion, está no cambiará a diferencia del anterior grafico 


```python
np.random.seed(0)
# Crearemos una lista de arrays donde voy a tener información de edades 
ages = [np.random.normal(30,5,100), np.random.normal(40,5,100), np.random.normal(35,5,100)]
```

De la misma forma ahora definiremos nuestro boxplot con matplotlib y en este caso recibe los siguientes parámetros: 
- La información a graficar 
- patch_artist, hace referencia a que las cajitas estén rellenas de color, este parámetro contiene un booleano, true o false. 
- Notch, recibe un booleano, añade un recorte para que podamos estar conscientes de la confianza que estamos estableciendo a los datos… es más una herramienta visual, si lo dejamos en falso nos mostrará cada caja cuadrada y sin cortes.
- vert, haciendo referencia a la orientación del grafico, recibe un booleano, si es false será el gráfico horizontal, los ejes cambian en automático. 
- Finalmente establezco los labels, fíjate que tenemos 3 grupos de datos, entonces ingresamos 3 nombres a nuestros datos 

```python
plt.boxplot(ages, patch_artist=True, notch=True, vert=True, labels=['Grupo 1', 'Grupo 2', 'Grupo 3'])

# Repetimos nombrar nuestros elementos y mostramos.
```

# Series de tiempo y manejo de fechas con Matplotlib
Digamos que queremos monitoriar la  producción diaria de energía solar para optimizar el uso de recursos y aumentar la eficiencia.
No es facil identificar patrones o anomalias al analizar simplemente las tablas.
Al crear un gráfico de lineas que muestre la producción de energía a lo largo del tiempo puedes detectar rápidamente las tendencias de consumo estacionales como el aumento de producción en verano y disminución en invierto por ejemplo.

La visualización de series temporales sirve para identificar patrones, tendencias y tambien anomalias PERO en los datos a lo largo del tiempo, estos no solo se aplican en el ámbito financiero si no que tambien en sectores de meteorología, economía y análisis de tráfico web e incluso salud.

Poder visualizar y representar datos temporales ayuda a anticipar cambios futuros y accionar proactivamente para la toma de decisiones. 

Para este ejemplo usaremos Numpy, Pandas y Matplotlib pues manejaremos un formato de fechas aleatorias en dos ejemplos aislados con la función "DateFormatter" del MÓDULO de matplotlib "DATES".

```python
import pandas as pd
from matplotlib.dates import DateFormatter

dates = pd.date_range(start='2023-01-01', periods = 100)
# A partir de que fecha quiero este rango de fechas
# Cuantos datos creará

values = np.random.rand(100).cumsum()
# Generación aleatoria de 100 valores
# Queremos hallar la suma acumulada para que este sea el tamaño de la serie

# DATA será el taframe que contendrá tanto las fechas como valores generados
data = pd.DataFrame({
    'Date' : dates,
    'Values' : values
})
```

# Comenzamos la graficación - gráfico de líneas, diferente sintaxis
Analicemos nuestro formato de fechas, si lo dejamos así dentro de nuestro gráfico las fechas se mostrarán sobreescritas, amontonadas unas con otras.

Lo que podemos hacer es rotar esta información, para ello usamos plt.xticks(rotation=int)

Este ejemplo es muy genérico, el sigueinte lo haremos aplicado a un caso.

Tenemos la representación de numeros en forma de suma acumulada, como va sumando cada número lo que vamos a ver en la serie de tiempo es que este va a ir aumentando con esto damos formato a una serie de tiempo.

```python
# Tendrémos dos objetos
# SUBPLOTS, esta en plural
fig, ax = plt.subplots(figsize=(12,6))

#USO DE MI SEGUNDO OBJETO de arriba.
#Defino aquí el tipo de gráfico que quiero.
# Fijate que la forma de pasar la información que utilizaré es distinta su sintaxis, no simplemente indico la varaible.
ax.plot(data['Date'], data['Values'], color = 'green')

plt.xticks(rotation=45)

plt.title('Serie de tiempo con formato en las fechas')
plt.xlabel('Date')
plt.ylabel('Value')
```

Para la variable date, donde creamos nuestras fechas de información, recuerda que esto lo puedes hacer de manera manual ya sea en un Numpy array o en un Dataframe de Pandas, en este caso para el ejemplo queremos que se creen de forma automática.

```python
# Creamos nuevamente la información de las fechas
date = pd.date_range(start='2023-01-01', periods=12, freq='M')
# En el segundo parámetro especifico la secuencia, voy a generar 12 fechas, una por cada mes, para tener en total las ventas desde enero hasta diciembre.

sales = np.random.randint(1000, 5000, 12)

# Creamos nuestra infromación de ventas en un DataFrame con Pandas...
sales_data = pd.DataFrame({
    'Date' : date,
    'Sales' : sales
})

#QUEREMOS UN GRÁFICO DE LINEAS
plt.plot(sales_data['Date'], sales_data['Sales'], marker='o', linestyle='-', label='Ventas mensuales')
# Pondremos nuestra marca en cada coordenada que se encuentre con mark
# Todo esto fue la personalización

# Observa que estamos manejando un formato en número para los xticks, esto podemos personalizarlo a formato fecha.
plt.gca().xaxis.set_major_formatter(DateFormatter('%b %Y'))
#Especificamos que queremos hablar del eje X en específico y lo valos a configurar

#INTEGRACIÓN DE MÉTODOS
plt.xticks(rotation=45)
plt.title('Análisis de ventas mensuales')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.legend()

plt.tight_layout()

plt.show()

#El formato de fecha que tenemos es en números pero podemos darle una personalización especificando el formato de FECHA, esto lo pondremos antes de la rotación.
```

# Varios gráficos en una sola figura
En muchas situaciones será necesario presentaar múltiples gráficos en una sola figura para comprar múltiples datos de manera eficiente.
Usaremos SUBPLOTS y LAYOUTS avanzados de Matplotlib...

# Gridspec
En lugar de crear variso ejemplos, los consolidaremos en un solo ejemplo con esta herramienta que permite crear layouts personalizados con subplots de diferentes tamaños y formas dentro de la misma figura.

Seguiremos usando matplotlib, numpy y un módulo en específico llamado gridspec

```python
import matplotlib.gridspec as gridspec

#Vamos a graficar dos veces el seno y crearemos un histograma, todo esto lo vamos a conglomerar de gridspec, grid - cuadrícula y spec de specification 

x = np.linspace(0,10, 100)
# Lamamos o creamos un arreglo de números igualmente espaciados... como parámetros 0 es el valor inicial de la secuencia y el valor final es el 10 y el último 100, es el número total de valores que se crean entre el 0 y 10

y = np.sin(x)
# Le estamos diciendo que Numpy genere el seno con los valores de x 

#Creamos un set de datos para nuestro histograma

data= np.random.randn(100) #Quero 100 datos

#Creamos la cuadrícula
gs = gridspec.GridSpec(2,2,height_ratios=[2,1], width_ratios=[1,1])
# Le pasamos primero que la grilla este dividida en dos columnas y en dos filas 
# Establecemos las proporciones tanto de la altura de las filas en la cuadricula como la parte de abajo con height_ratios=[2,1]
# Le pasamos widht_ratios, ambas columnas van a tener el mismo ancho para este caso, de la misma forma le pasamos una lista

#PODEMOS DAR MAYOR PERSONALIZACIÓN
fig = plt.figure(figsize=(10,8))

#Queda inclustar los gráficos que hemos creados una vez teniendo la información y personalización, recuerda que ya elegiste 2 gráficos, en este caso tendremos un gráfico en grande y debeajo de el 2 más.

# Primer subplot grande, ocupa toda la primera fila, con esto estamos estableciendo su posición
ax1 = fig.add_subplot(gs[0,:]) # Va en la fila 0, de inicio a fin, eso indicamos
#Luego vamos a graficar tanto x como y
ax1.plot(x,y, color='blue')
ax1.set_title('Graficación de Seno de X')
ax1.set_xlabel('x')
ax1.set_ylabel('Sen(X)')

#SEGUNDOS SUBPLOT, ocupa la esquina inferior a la izquierda, creación del segundo OBJETO
ax2 = fig.add_subplot(gs[1,0]) #Imaginalo como una matriz
ax2.hist(data, bins=20, color='purple', edgecolor='black')
ax2.set_title('Histograma')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Frecuencia')

#Tercer subplot en la esquina inferior derecha
ax3 = fig.add_subplot(gs[1,1]) #Imaginalo como una matriz, su índice 
ax3.scatter(x,y,color='red')
ax3.set_title('Dispersión de Seno')
ax3.set_xlabel('x')
ax3.set_ylabel('sin(x)')

#Ajustar el layout para evitar la superposición de las figuras
plt.tight_layout()

plt.show()
```
