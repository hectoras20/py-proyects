# Cálculo de Prima ÚNICA de Riesgo para SEGURO DE VIDA

\[
A_{\overset{1}{x}:n} = \sum_{t=t}^{n-1} {}_tP_x \cdot q_{x+t} \cdot v^{t+1}

Llamemos las funciones que nos ayudarán a nuestros cálculos, habrá que modificarlas para que sirvan de forma dinámica y no estática
- Para obtener los valores t_P_x de forma recursiva 
- Para obtener los valores Q_x adaptada para servir de forma general

- Posteriormente habría que crear una función para otro tipo de tasas
- Añadir una función de valor presente actuarial sobre sobrevivencia, este podría servirnos posteriormente en seguros dotales puros.
- Con la anterior funcion de VPA podría definir de una manera más fácil la función de Anualidades temporales anticipadas pues podría aplicar un ciclo o recursividad para obtener este concepto con base al VPA, pues podríamos verlo en código como  la suma de n-1 VPAs.

```python
# Para obtener los valores t_P_x de forma recursiva
def CALCULAR_t_P_x(t, P_xt):
    if t == 0:
        return 1
    else:
        return CALCULAR_t_P_x(t - 1, P_xt) * P_xt[t - 1]

    
# Para obtener los valores Q_x adaptada para servir de forma general
def CALCULAR_Q_x(t, edadContrata, tabla):
    buscar_xt = edadContrata + t
    valor = tabla.loc[tabla['Edad']  == buscar_xt,'q_x'].item()

    print(f'El valor de q_{buscar_xt} es {valor}')
    return valor

print(CALCULAR_Q_x(5, 20, CNSF2013M))



```

```python
def PUR_vida(tasa, edadContrata, SA, tabla, tiempoSeguro = 1):
    # Defino mis herramientas
    # Valores que tomará t dentro del flujo:
    valoresT = np.arange(tiempoSeguro)
    # Valores que tomará X dentro de los flujos:
    valoresX_edades = edadContrata + valoresT
    # Obtenemos por valores de P_x+t - ESTOS VALORES NOS SIRVEN PARA LA PARTE DE RECURSIVIDAD
    P_xt = []
    for i in valoresX_edades:
        P_xt.append(tabla.loc[tabla['Edad']  == i,'p_x'].item())

    # Función para obtener los valores t_P_x
    list_t_P_x = []
    for i in valoresT:
        list_t_P_x.append(CALCULAR_t_P_x(i, P_xt))
    # Convertimos a un numpy array para facilitar flujos
    t_P_x = np.array(list_t_P_x)

    lista_Q_x = []
    # Obtener valores Q_x
    for i in valoresT: # Recordemos que dentro de esta función se suma la edad con los valores que tome t... por ende usamos valoresT pero hay más formas de hacerlo como comentabamos, pues pudimos pasar valores en donde la edad y t ya esten sumados, por ejemplo. (pudimos usar la variable "valoresX_edades")
        lista_Q_x.append(CALCULAR_Q_x(i, edadContrata, CNSF2013M))
    #Convertimos esta lista a un numpy array para facilitar los flujos
    Q_x = np.array(lista_Q_x)

    # Ahora vamos con el valor presente de cada intervalo, de cada flujo
    listaVt = []
    for i in valoresT:
        listaVt.append((1/(1+(tasa/100)))**i)
     #Convertimos esta lista a un numpy array para facilitar los flujos
    Vt = np.array(listaVt)

    flujoContingente = SA * t_P_x * Q_x * Vt
    PUR = float(flujoContingente.sum())

    #Dentro de una tupla nos retornará lo siguiente
    return flujoContingente, PUR


SeguroVida1 = PUR_vida(3.5, 20, 50000, CNSF2013M, 5)
print('Flujo Continente =', SeguroVida1[0],'\nPRIMA ÚNICA DE RIESGO = ',SeguroVida1[1])

    

    


    
    
```

# PRIMA NETA NIVELADA (opcional)
La PNN la definimos dentro de clases como la PUR/Anualidad Anticipada Temporal

**COBERTURA DE RIESGO / NUMERO DE PAGOS CONTINGENTES***

Claramente la PNN debe de ser menor o igual a la temporalidad de cobertura del seguro, en este caso, al no ser vitalicio, nos centraremos en una temporalidad definida.
Es anticipada porque como tal, fuera del tiempo de gracia, las primas se pagan primero para poder acceder al servicio de seguro.

## ¿Cómo se definía una Anualidad Anticipada?
- No es más que una suma de flujos desde t=0 hasta n-1, arranca desde t=0 por ser anticipada, n-1 por la cobertura que alcanza el seguro siendo este su último intervalo de cubrimiento.
- Los flujos son t_P_x * v^t

En este caso, definiremos una función para ese calculo de anualidades, este sería otro método calculador que realizaremos, de inicio, podría servirnos las anteriores funciones que ya definimos de búsqueda de valores. 

**Veamos que necesitamos valores que definimos dentro de la función que calcula la PUR, como lo es la TASA, EDAD Y TABLA**, sin embargo, tengo la confianza que dentro de la clase, estas variables se pueden relacionar directamente con índices, pues podría ser esto un candidato de clase.

OJO: la t tiene que ser menor o igual al tiempo de cobertura del seguro, o bien, dentro de nuestro codigo, tiene que ser menor o igual a la t que se usó para la función de calculo de PUR

```python
def PNN(PUR, edad, tasa, tabla, temporalidad = 1):
    # Queremos t_P_x, existe una función que definimos globalmente
    # La temporalidad "t" NO ES LA MISMA CON LA QUE SE CALCULÓ la PUR, esta se define nuevamente y tiene que ser menor o igual a la t de la PUR
    pnn_T = np.arange(temporalidad)

    # Para poder usar nuestra funcion que calcula t_P_x necesitamos previamente los valores de P_x+t, estos no son más que una búsqueda con LOC dentro de nuestro dataframe PERO, necesitamos ademas definir las edades con las que se trabaja P_x+t osea definir una lista y obtener los valores "x+t"
    valores_xt = edad + pnn_T

    # Ahora si, podemos usar nuestras edades "x+t" para que se haga la búsqueda y se almacenene dentro de una lista
    P_xt = []
    for i in valores_xt:
        P_xt.append(tabla.loc[tabla['Edad']  == i,'p_x'].item())

    list_t_P_x = []
    for i in pnn_T:
        list_t_P_x.append(CALCULAR_t_P_x(i, P_xt))
    # AL SER VALORES PARA NUESTRO FLUJO, NECESITAMOS PASARLA A UN NUMPY ARRAY
    t_P_x = np.array(list_t_P_x)

    # Ahora necesitamos la otra parte de nuestro flujo, los valores presentes
    listaVt = []
    for i in pnn_T:
        listaVt.append((1/(1+(tasa/100)))**i)
     #Convertimos esta lista a un numpy array para facilitar los flujos
    Vt = np.array(listaVt)

    flujoAnualidad = Vt * t_P_x
    valor_Anualidad = float(flujoAnualidad.sum())

    return (PUR/valor_Anualidad)

print(PNN(SeguroVida1[1], 20, 3.5, CNSF2013M, 5))
    
```

# PRIMA DE TARIFA
La prima de tarifa se definió como la Prima de Riesgo (sea PUR o PNN) / ( 1 - %alpha)
Donde *%alpha* son los recargos () que se aplican dentro del seguro y que corresponden a 3 tipos de gastos:
- Gastos por adquisición
- Gastos de administración y/o operación
- Margen de Utilidad

En donde cada uno puede tener diferente durabilidad durante la temporalidad del seguro, por ejemplo:

- GASTOS DE ADQUISICIÓN - A los agentes se les paga dichos gastos, sin embargo, el pago de estos es en un tiempo limitado, ya que no les es grato que se les pague durante un periodo largo de tiempo. *k<=t - depende de la temporalidad del seguro*
Por ejemplo en un seguro ordinario de vida = vitalicio por fallecimiento, se paga por un periodo k

- GASTOS DE ADMINISTRACIÓN - La aseguradora **debe cobrar durante todo el plazo o vigencia del seguro**. *k=t*

- La empresa de seguros **puede cobrar durante toda la vigencia del seguro o en un periodo menor al plazo o vigencia del seguro**. *k<=t*


Osea que la temporalidad k que dure o que tenga cada gasto es importante, pues definirá el porcentaje que se le recargará por cada gasto, se trabaja con la temporalidad que dure, y se obtienen los porcentajes que se aplicarán sobre los calculos.


**El calculo** de los porcentajes de recargo que se restan en el denominador siguen una estructura similar...

```python
# La suma asegurada es fija PERO CON ESTE CODIGO TIENE LA OPCIÓN DE PASARLA COMO UN NUMPY ARRAY Y SEGUIR FUNCIONANDO

def new_PUR_vida(tasa, edad, SA, tabla, tiempoSeguro = 1):
    # SE CALCULA BAJO UNA SUMA, DEFINAMOS LOS VALORES DE t 
    valoresTF = np.arange(tiempoSeguro)
    # Debe de haber "tiempoSeguro -1" valores

    valoresFlujot_P_x_vt = [] # ESTO ME DARÍA DE MI SUMATORIA LA PARTE DE t_p_x * v^t
    valoresFlujoQ_x = [] # ESTO ME DARÍA DE MI SUMATORIA LA PARTE DE q_x
    
    for i in valoresTF:
        valoresFlujot_P_x_vt.append(VPA(i,edad,tabla,tasa)) # RECUERDA QUE ESTO SOLO ME RETORNA UN FLUJO, el flujo t_P_x * v^t UNICAMENTE y necesito n-1 flujos
        valoresFlujoQ_x.append(CALCULAR_Q_x(i, edad, tabla))

    # Pasamos a un numpy array
    flujot_P_x_vt = np.array(valoresFlujot_P_x_vt)
    flujoQ_x = np.array(valoresFlujoQ_x)

    flujoTotal = flujot_P_x_vt * flujoQ_x
    valorFlujo = float(flujoTotal.sum())

    return SA * valorFlujo


SeguroVida2 = new_PUR_vida(3.5, 20, 50000, CNSF2013M, 5)
print(SeguroVida2)
```

```python
# Neceistamos que el usuario nos dé las tasa con las cuales se aplicarán en gastos
# Esta lista se usa para el calculo de Gadm y Prima de Tarifa dentro de la función
tasas_Admin = [4,3,2]
# Estas listas se usan para el calculo de Prima de tarifa
tasas_Adqui = [10,5,2]
tasas_MG = [4,3,2]


def BEL(edad,tabla, tasaPoliza, SA, tasas_Admin = list, tasas_Adqui = list, tasas_MG = list, tiempoVigenciaPoliza = int, numPagosADQ = int, numPagoPrimas=int):
    # Creamos una lista que contendrá los valores de t, desde 0 hasta t, esto debido a que los calculos sobre cada periodo en el que queramos calcular el BEL van cambiando
    t_vigenciaPoliza = np.arange(tiempoVigenciaPoliza)

    # Creamos una para el tamaño de la prima
    t_vigenciaADQ = np.arange(numPagosADQ)

    tasaPoliza = tasaPoliza/100

    #Imprimiremos como se ve la función y que pasos realiza internamente
    print('Elementos de construcción - Se puede omitir e ir al siguiente bloque de información')

    # 1. Gastos de administración NIVELADOS con la vigencia del seguro
    # Mientras corre en t este no cambia SE MANTIENE FIJO
    Gadm = CALCULADOR_per_gastos(edad,tabla,tasaPoliza, tiempoVigenciaPoliza, tasas_Admin) # Me retorna un float

    print('Gastos Adm. =', Gadm)

    # 2. Calculo de egresos futuros con base al seguro en cuestión
    # Este si va cambiando confirme corre t, POR ENDE SE ME OCURRE GUARDAR SUS VALORES EN UN NUMPY ARRAY
    lista_dot = []
    # Aplicaremos un reversed creo que por comodidad y no tener problema con los calculos, LO UNICO QUE CAMBIARIA ES EL ORDEN EN EL CUAL SE DAN LOS DATOS
    for i in reversed(t_vigenciaPoliza+1):
        # Aplicamos el calculo de la PUR del seguro en cuestión, en este caso es un seguro temporal de vida, tenemos 3 funciones que reciben los mismos parámetros pero que estan construidas de diferente forma, si algo está mal en la aplicación de conceptos en relación a MASPI, solo es cuestión de cambiar el nombre de la función empleada.
        lista_dot.append(seguro_fallecimiento(i, edad, tabla, tasaPoliza)) # Retorna un FLOAT
    valores_dot = np.array(lista_dot) * SA

    PUR = shortest_PUR_vida(tasaPoliza, edad, SA, tabla, tiempoVigenciaPoliza)
    PNN = new_PNN(PUR, edad, tasaPoliza, tabla, numPagoPrimas)
    # 3. VPE por gastos de administración, ocuparemos: función de ANUALIDAD bajo la temporalidad de VIGENCIA DE SEGURO, La variable Gadm y la función de Prima de Tarifa
    # LA PRIMA DE TARIFA ES CONSTANTE DENTRO DE LOS CALCULOS
    PT = new_PT(PNN, edad, tabla, tasaPoliza, numPagoPrimas, tasas_Admin, tasas_Adqui, tasas_MG)


    # La anualidad SI VA CAMBIANDO
    # CUIDEMOS LA LONGITUD, TAMAÑO DE NUESTRAS LISTAS, hasta ahorita no hemos creado ninguna lista con diferente dimensión a la de la vigencia del seguro
    # El siguiente código pudimos haberlo hecho arriba, pero para temas educativos y de orden de acuerdo a las notas publicadas, lo haremos así
    adm = t_vigenciaPoliza + 1
    lista_anualidades_adm = []
    for i in reversed(adm):
        lista_anualidades_adm.append(anualidad_Temp_Anticipada(tabla, tasaPoliza, i, edad))
    
    anualidades_administracion = np.array(lista_anualidades_adm) # Recuerda que debe de haber una consistencia con los valores, el priimer valor es de t=4, el segundo de t=3 y así hasta 0 

    # Calculamos el VPE por gastos de administración:
    # VALOR * TAMAÑO 5 * TAMAÑO 5
    VPE_por_gastos_administracion = Gadm * PT * anualidades_administracion
    # ESTO NOS RETORNA UNA LISTA, UN NUMPY ARRAY DE DIMENSIÓN VIGENCIA DE POLIZA

    # 4. LOS SIGUIENTES ELEMENTOS SERÁN NIVELADOS BAJO EL PLAZO DE PAGOS DE GASTOS DE ADQUISICIÓN: Anualidad y Gadq 
    adq = t_vigenciaADQ +1
    lista_anualidades_adquisicion = []
    for i in reversed(adq):
        lista_anualidades_adquisicion.append(anualidad_Temp_Anticipada(tabla, tasaPoliza, i, edad))

    anualidades_adquisicion = np.array(lista_anualidades_adquisicion)

    #Ahora obtendremos el valor de los Gastos adquisitivos
    # ESTE ES CONTINUO, FIJO SOBRE ESTE FLUJO
    Gadq = CALCULADOR_per_gastos(edad, tabla, tasaPoliza, numPagosADQ, tasas_Adqui) # Me retorna un float

    # Entonces... si tratamos de multiplicar la PT * Gadq * anualidades_adquisicion ESETO ES DE LA FORMA = TAMAÑO 5 * VALOR * TAMAÑO 3!!!! 
    # CAMBIERON ESE TAMAÑO DE 3 A 5, para ele ejmplo del clasrrom, pero este se generaliza asi:
    dif_adq = len(t_vigenciaPoliza) - len(anualidades_adquisicion)
    if dif_adq>0:
        anualidades_adquisicion = np.concatenate([anualidades_adquisicion, np.full(dif_adq, False)]) # Se usará para multiplicar por ende se usa True = 1
    
    # AHORA SI, TAMAÑO 5 * VALOR * TAMAÑO 5 = TAMAÑO 5
    VPE_por_gastos_adquisicion = PT * Gadq * anualidades_adquisicion


    # SEGUIMOS CON EL CALCULO DE VALOR PRESENTE DE INGRESOS
    # ESTE SI SE CALCULA BAJO EL PLAZO DE PAGO DE PRIMAS
    t_primas = np.arange(numPagoPrimas) +1
    edades = np.array(edad) + t_primas # 40, 41, 42

    lista_primasTarifa = [] 

    for i, edad_actual in zip(reversed(t_primas), edades): 
        lista_primasTarifa.append(anualidad_Temp_Anticipada(tabla, tasaPoliza, i, edad_actual))

    anualidadPARAprimas_de_tarifa = np.array(lista_primasTarifa)

    # TAMAÑO 5 * TAMAÑO 3! = PT * anualidadPARAprimas_de_tarifa
    # CAMBIAMOS ESO
    dif = len(t_vigenciaPoliza) - len(anualidadPARAprimas_de_tarifa)
    if dif>0:
        anualidadPARAprimas_de_tarifa = np.concatenate([anualidadPARAprimas_de_tarifa, np.full(dif, False)])

    # AHORA SI ESTAN EN EL MISMO TAMAÑO (5)
    VPI_por_primas_tarifa = PT * anualidadPARAprimas_de_tarifa
    

    BEL = valores_dot + VPE_por_gastos_administracion + VPE_por_gastos_adquisicion - VPI_por_primas_tarifa
    # ESTO RETORNA UN ARRAY CON LAS BEL EN LOS TIEMPOS T DE LA VIGENCIA DE UN SEGURO TEMPORAL DE VIDA

    #Veamos que valores retorna con respecto a t=4 osea el primer año
    for i in t_vigenciaPoliza:
        print(f'Los calculos en tiempo t=i se ven así: VPE_fall = {valores_dot[i]} + VPE_adm = {VPE_por_gastos_administracion[i]} + VPE_adq = {VPE_por_gastos_adquisicion[i]} - VPI = {VPI_por_primas_tarifa[i]}')


    return BEL

print(BEL(40, CNSF2013M, 6.2, 100000, tasas_Admin, tasas_Adqui, tasas_MG, 5, 3, 3))







```
